# EDA  Exploratory Data Analysis

```python
EDA is the data exploration technique to understand the various aspects of the data.

It includes several steps or technique in sequence that we need to follow.

We must ensure that the data must be clean and does not have data redundancy, any missing or null values ,
 and any duplicate values in the dataset.

We must also identify the important variables in the data sets and remove all the unnecessary unwanted values in the data 
that may hinder the accuracy of our conclusions when we work on those datas model building.

Also we must underatand relationships b/w variables through EDA.

At last we must be able to derive conclusions on gathering insights about the data for the conclusive interpretations
in order to move on to more complex processes in the data processing life cycles.
```

# Objective of EDA in Data exploration

```python
It is basically used to filter the data from redundencies.

1. EDA helps in identifying the faulty points in the Data. Remove and then clean your Data.

2. EDA helps us to understand the relationship b/w the variables which gives us the wider perspective on the
   data and it actually helps us build on it by utilizing the relationship b/w the Variables.
```

# Steps Involved in EDA

```python
1. Understand the DATA : understand the Variables in the Data Set What kind of Data You are working with ,
   what kind of Rows and Columns you are working with.

2. Clean the DATA : Clean the Data from the Redundancies which can be irregularity in the data ,
    unnecessary Columns  that are not required for making our conclusions or interpretations. 
    Or the Outliers that can cause Noise in the data that can underfit or overfit the model when we are working on the
    model Building as well.
    
3. Analyse the relationship b/w the Variables      
```
