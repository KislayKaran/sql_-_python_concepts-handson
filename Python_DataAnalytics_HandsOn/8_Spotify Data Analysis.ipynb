```python
# setting 'release_date' as the index column in the main dataframe

df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)     # changing  into datetime format
df_tracks.head()
```

```python
#index location to filter specific rows in the dataframe

df_tracks['artists'].iloc[18]
```

# Converting the column duration_ms into seconds

```python


df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))  
#using the lambda function  to divide it by 1000

df_tracks.drop("duration_ms", inplace= True, axis=1)

df_tracks.duration.head()
```

# Correlation Mapping  by applying Pearson Correlation method

```python

 
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")   #Dropping 3 unwanted Columns
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap b/w Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)


# annot= True, this will write the data values in each cell
# fmt=".1g", this is string formating code used when adding annotations
# vmin ,vamx values to anchor the color map otherwise they are inferred from the data and other keyword arguments.
# cmap  color map can select from the palette
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

# Regression plot b/w loudness and energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color= "c").set(title="Loudness Vs Energy Correlation")
```

# Regression plot b/w popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color= "b").set(title="Popularity Vs  Acousticness")
```

# Creating a new column 'year'

```python


df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

# Distribution plot for the total songs released each year

```python


sns.displot(years,discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per Year")
```

# Duration of songs over the years

```python


total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x= years, y=total_dr, ax=ax, errwidth= False).set(title= "Year Vs Duration")
plt.xticks(rotation=90)
```

# Lineplot for the Avg. duration of the songs

```python

total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year Vs Duration")
plt.xticks(rotation=60)
```
