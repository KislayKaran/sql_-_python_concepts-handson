```python
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
pl.plot(x,a,'b')
pl.plot(x,b,'r')
pl.show()
```

```python
x=np.arange(0.,10,0.1)
p=np.cos(x)
q=np.sin(x)
pl.plot(x,p,'c',linewidth=2,linestyle='dashdot')
pl.plot(x,q,'m',linewidth=4,linestyle='dashed')
pl.show()
```

```python
ar2=[1,7,21,35,35,21,7,1]
s2=np.sin(ar2)
c2=np.cos(ar2)
t2=np.tan(ar2)
pl.figure(figsize=(9,5))
pl.plot(ar2,s2,'c')
pl.plot(ar2,c2,'r')
pl.plot(ar2,t2,'k',linestyle="dotted")
pl.xlabel('Array Values')
pl.ylabel('Sine, Cosine and Tangent Values')
pl.show()
```

(C) Changing Marker Type, Size and Color

    marker = <marker type> , markersize = <in points> , markeredgecolor = <valid color>
    
'.'  point marker     's'  square marker          'o'  circle marker     '*'  star marker         '+'  plus marker 

'D'  diamond marker   'd'  thin_diamond marker    'h'  hexagon1 marker   'H'  hexagon2 marker     'x'  x marker

'|'  vline marker     '_' hline marker 

NOTE : You can combine the marker type with linecolor code 

      e.g., 'r+' line color as 'red' and marker type as '+'.

```python
m=[1,2,3,4]
n=[2,4,6,8]
pl.plot(m,n, marker='D',markersize=5, markeredgecolor= 'red')
```

```python
pl.plot(m,n,'r+',linestyle="solid")      #Line color and marker style combined so marker takes the same color as line.
```

```python
pl.plot(m,n,'r+',linestyle="solid",markeredgecolor='b')  #Here marker color seperately specified.  
```

```python
pl.plot(m,n,'ro')     #If  you don't specify the linestyle argument seperately along with linecolor & markerstyle-string 
                      #Python will only plot the markers and not the line.
```

# Scatter Charts

# Using plot( ) function

If you specify the 'linecolor' and 'markerstyle'-string without the 'linestyle' argument, then the plot created resembles 
a scatter chart as only the datapoints are plotted now.

In plot() function, whenever you specify marker type/style, whether with color or without color,
and do not give 'linestyle' argument, plot() will create a scatter chart.

```python
import numpy as np 
import matplotlib.pyplot as pl
a=np.arange(1,20,1.25)
b=np.log(a)
pl.plot(a,b,'ro')
pl.xlabel('Random Values')
pl.ylabel('Logarithm Values')
pl.show()
```

```python
a=np.arange(1,20,1.25)
b=np.log(a)
c=np.log10(a)
pl.plot(a,b,'ro')
pl.plot(a,c,'b^')
pl.xlabel('Random Values')
pl.ylabel('Logarithm Values')
pl.show()
```

# Scatter Charts

# Using scatter( ) function

matplotlib.pyplot.scatter(<array1>,<array2>)

or

<pyplot aliasname>.scatter(<array1>,<array2>)
    e.g. pl.scatter(a1,a4)                      # a1 & a4 are nd arrays
    
         pl.scatter(a1,a4, marker="x")

Specifying size of the Markers  (argument s of scatter())

Specifying color of the Markers (argument c of scatter())


pl.scatter(x,y,s=None, c= None, marker= None)

x,y     : The data positions
  s     : The Marker size 
  c     : Marker color, sequence, or sequence of color, optional argument.
marker  : MarkerStyle

Specifying Varying Colors and Sizes for Data Points

we need to specify an array  of colors having the same shape as arrays being plotted (x & y)  as the value of c argument and
an array of sizes having the same shape as arrays being plotted (x & y) as the value of s argument.

When c or s argument have a sequence/array with multiple color/size values,
then each color/size is applied to different data points in order,
i.e., first color and size to the first data point ,second color and size to the second data point.

```python
a1=np.linspace(-1,1,5)
a2=np.exp(a1)
colarr=['r','b','m','g','k']
sarr=[20,60,100,45,25]
pl.scatter(a1,a2,c=colarr,s=sarr)   # 'c' argument has sequence of colors; each color will be assigned to different data points.
                                    # 's' argument has sequence of sizes ; one size for each data point
```

```python
import numpy as np
import matplotlib.pyplot as pl
x1=np.random.randint(1,100,size=(250,))
y1=np.random.randint(1,100,size=(250,))
size=range(1,60,5)
colr=['r','b','c','m','g','k']
pl.scatter(x1,y1, c=colr, s=size)
pl.xlabel('X Values')
pl.ylabel('Y Values')
pl.show()
```

```python
import numpy as np
import matplotlib.pyplot as pl
x1=np.random.randint(1,100,size=(250,))
y1=np.random.randint(1,100,size=(250,))
z1=np.random.randint(1,100,size=(250,))
size=range(1,60,5)
pl.scatter(x1,y1, color='r', s=size)
pl.scatter(x1,z1, color='r', s=size)
pl.xlabel('X Values')
pl.ylabel('Y & Z Values')
pl.show()
```
