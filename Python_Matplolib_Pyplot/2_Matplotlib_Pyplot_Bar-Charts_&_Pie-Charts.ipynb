# Creating Bar Charts 

Vertically or Horizontally using rectangles or bars of different heights/widths . 
Pyplot offers bar() function to create a bar chart where you can specify the sequences for x-axis and 
corresponding sequence to be plotted on y-axis.
each value is plotted as bar on corresponding x-value on x-axis.

```python
import matplotlib.pyplot as pl
a,b,c=[1,2,3,4],[2,4,6,8],[1,4,9,16]
pl.bar(a,b)
pl.xlabel('Values')
pl.ylabel('Doubles')
```

```python
pl.bar(a,c)
pl.xlabel('Values')
pl.ylabel('Squares')
```

```python
Cities=['Delhi','Mumbai','Bangalore','Hyderabad']
Population=[23456123,20083104,18456123,13411093]
pl.bar(Cities,Population)
pl.xlabel('Cities')
pl.ylabel('Population')
pl.show()
```

NOTE : The order of bars plotted may be different from the order in actual data sequence.
       Preffered order is in the Alphabetical order.

```python
import matplotlib.pyplot as pl
Info=['Gold','Silver','Bronze','Total']
Australia=[80,59,59,198]
pl.bar(Info,Australia)
pl.xlabel("Medal Type")
pl.ylabel("Australia Medal count")
pl.grid(True)
pl.show()
```

```python
import matplotlib.pyplot as pl
Info=['Gold','Silver','Bronze','Total']
Australia=[80,59,59,198]
India=[26,20,20,66]
pl.bar(Info,Australia)
pl.bar(Info,India)
pl.xlabel("Medal Type")
pl.ylabel("Australia,India Medal count")
pl.grid(True)
pl.show()
```

# Changing Widths of the bars in a Bar Chart 

You can specify  a different  width  ( other  than the default width ) for all the bars of a bar chart.

    matplotlib.pyplot.bar(x-sequence,y-sequence, width= <float value>)
    
You can also specify different widths for different bars of a bar chart.

    matplotlib.pyplot.bar(x-sequence,y-sequence, width= <width values sequence>)

```python
pl.bar(Cities,Population,width=0.3)
```

```python
pl.bar(Cities,Population,width=[0.3,0.5,0.6,0.8])
```

# Creating Multiple Bar Charts

(i) Decide number of X points. Firstly, you need to determine how many X points you will need.
For this, calculate the number of entries in the ranges being plotted. 

Say, you are plotting two sequences A and B, so length of sequences A or B
being plotted will determine the number of X points in our case.

NOTE : ** You can create a sequence with this length of A or B using range() or numpy.arange() functions.
Do note that both the sequences A and B must have similar shape i.e., same length (number of elements).

(ii) Decide thickness of each bar and accordingly adjust X points on X-axis.
So, if you want to plot two bars per X point, then carefully think of the thickness of each bar.

Say you want the thickness of each bar as 3, then for first range keep it as X and for the second data range,
make the X point as X + 0.3. (see below)

(iii) Give different color to different data ranges using color argument of bar() function.

(iv) The width argument remains the same for all ranges being plotted.

(v) Plot using bar() for each range separately, i.e., the number of bars decide the number of bar() functions
you will be using to plot different bars on same plot. 

```python
import numpy as np
import matplotlib.pyplot as pl
pl.figure(figsize=(10,7))
val=[[5.,25.,45.,20.],[4.,23.,49.,17.],[6.,22.,47.,19.]]
X=np.arange(4)
pl.bar(X+0.00,val[0],color='r',width=0.25)
pl.bar(X+0.25,val[1],color='g',width=0.25)
pl.bar(X+0.50,val[2],color='b',width=0.25)
pl.grid(True)
pl.show()
```

```python
import numpy as np
import matplotlib.pyplot as pl
Info=['Gold','Silver','Bronze','Total']
pl.figure(figsize=(10,7))
Australia=[80,59,59,198]
England=[45,45,46,136]
India=[26,20,20,66]
Canada=[15,40,27,82]
X=np.arange(len(Info))
pl.bar(X+0.00,Australia,width=0.15)
pl.bar(X+0.15,England,width=0.15)
pl.bar(X+0.30,India,width=0.15)
pl.bar(X+0.45,Canada,width=0.15)
pl.xlabel("Medal Type")
pl.ylabel("Australia,England,India,Canada Medal Count")
pl.grid(True)
pl.show()
```

# Creating a Horizontal Bar Chart

To create a horizontal bar chart , you need to use barh() function (bar horizontal), instead of bar().

```python
Cities=['Delhi','Mumbai','Bangalore','Hyderabad']
Population=[23456123,20083104,18456123,13411093]
pl.barh(Cities,Population)
pl.xlabel('Population')
pl.ylabel('Cities')
pl.show()
```

```python
Height=[5.1,5.5,6.0,5.0,6.3]
Names=('Adam','Bela','Chris','Diya','Sam')
pl.barh(Names,Height)
pl.xlabel("Height")
pl.ylabel("Names")
pl.show()
```

# Creating Pie Charts

Pyplot  interface offers pie() function for creating a pie chart.

(1) The pie() function, plots a single data range only. 
    It will calculate the share of individual elements of the data range being plotted vs the whole of the data range.

(ii) The default shape of a pie chart is oval but you can always change to circle by using axis() of pyplot,
     sending "equal" as argument to it. 
     That is, issue following command before you use pie() function :

         matplotlib.pyplot.axis ("equal")
     
         pl.axis("equal")



```python
import numpy as np
import matplotlib.pyplot as pl
contri=[17,8.8,12.75,14,9.7]
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,autopct="%1.1f%%")                 # The data range being plotted in pie()
                                                               # Labels for the Slices of pie
                                                               # Adding the formatted Slice Percentages to pie, using 'autopct'.
```

The format string used with autopct will deteremine the format of the percentage being displayed.
The format string begins with "%" operator, which specifies the format of the percentage value being displayed.

The general syntax for a format placeholder is
     
     "%[flags][width][.precision]type"

➜ The first % symbol is a special character that signifies that it is a 
special string which will determine the format of the values to be displayed.

➜  The width determines the total number of characters to be displayed 
(digits before and after decimal point + 1 for decimal point). 
    If the value being displayed has lesser number of digits than the width specified, 
    then the value is padded as per the flag specified. 
    If however, the value has more digits than the width specifies, then the full value is displayed.

➜ In pie charts, the most useful flag is 0 (zero), which when specified will pad the value being displayed
with preceding zeros if the digits of value is less the width.

➜ The precision is specified with digits and it specifies the number of digits after decimal point.
The type specifies the type of value: d or i means integer type and for F means float type.

➜ To print a % sign, given %% (two percentage signs) in the format string. 
The two %% signs are needed to print a percentage sign, otherwise single % is interpreted as a special character
with a special meaning and thus used accordingly. 
By doubling %, i.e., %%, you suppress its special meaning and thus a percentage sign is printed.

```python
contri=[17,8.8,12.75,14,9.7]
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,autopct="%3d%%")       #
```

```python
contri=[17,8.8,12.75,14,9.7]
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,autopct="%03d%%") 
```

```python
contri=[17,8.8,12.75,14,9.7]
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,autopct="%05.3f%%") 
```

```python
col=[8000,12000,9800,11200,15500,7300]
section=['A','B','C','D','E','F']
pl.title("Volunteering week Collectiion")
pl.axis("equal")
pl.pie(col,labels=section,autopct="%5.2f%%")
pl.show()
```

Changing Colors of the Slices

```python
contri=[17,8.8,12.75,14,9.7]
color=['red','cyan','pink','yellow','silver']
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,colors=color,autopct="%2.2f%%") 
```

Exploding a slice :

Sometimes in order to emphasize on one or more slices and show them little pulled out. This feature is called 'explode'

you just need to mention the unit distance at which that particular slice(s) is supposed to get explode 
in their respective order inside the  sequence.

```python
contri=[17,8.8,12.75,14,9.7]
color=['red','cyan','pink','yellow','silver']
expl=[0,0.2,0,0,0]
houses=['Vidya','Karishma','Namrata','Karuna','Sandy']
pl.pie(contri,labels=houses,explode=expl,colors=color,autopct="%2.2f%%") 
```

```python
col=[8000,12000,9800,11200,15500,7300]
section=['A','B','C','D','E','F']
expl=[0,0.2,0,0,0.3,0]
pl.title("Volunteering week Collectiion")
pl.axis("equal")
pl.pie(col,labels=section,explode=expl,autopct="%5.2f%%")
pl.show()
```
