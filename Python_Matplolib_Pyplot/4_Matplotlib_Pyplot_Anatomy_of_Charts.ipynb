# Anatomy of Chart

Any graph or Chart that we create using matplotlib's Pyplot interface is created as per specific structure of a plot.
Pyplot chaarts have hiearchial structures or in a simple words they are actually like containers containing multiple items/things inside it.

```python
1. Figure : to set the size of the plot. figsize 
    
    <matplotlib.pyplot>.figure(figsize=(<width>,<length>))
    
2. Axes  (Axis labels, Limits, Tick Marks)

    xlabel() & ylabel()
    
    xlim() & ylim()
    
    <matplotlib.pyplot>.xlim(<xmin>,<xmax>)    Sets the x-axis limits as xmin to xmax
    
    <matplotlib.pyplot>.ylim(<ymin>,<ymax>)    Sets the y-axis limits as ymin to ymax
    
    Ticks for Axes
    
    for x-axis : xticks(<sequence containing tick data points>,[<optional sequence containing tick labels])
    
    for y-axis : yticks(<sequence containing tick data points>,[<optional sequence containing tick labels])



3. Title : Text appearing at the top of the Plot  
    
    <matplotlib.pyplot>.title(<title string>)
    
4. Legends : These are different colors that identify different sets of data plotted.
    
    <matplotlib.pyplot>.legend(loc=<position number or string>)
    
    1 - 'upper right'
    2 - 'upper left'
    3 - 'lower left'
    4 - 'lower right'
    
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
X=np.arange(4)
Y=[5,25,45,20]
plt.xlim(-2.0,4.0)          # <matplotlib.pyplot>.xlim(<xmin>,<xmax>)
plt.bar(X,Y)
plt.title("A Bar Chart")    # <matplotlib.pyplot>.title(<title string>)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
X=np.arange(4)
Y=[5,25,45,20]
plt.xlim(-4.0,1.0)          # <matplotlib.pyplot>.xlim(<xmin>,<xmax>)
plt.bar(X,Y)
plt.title("A Bar Chart")    # <matplotlib.pyplot>.title(<title string>)
plt.show()
```

```python
q=range(4)
s=[23.5,25,26,28.5]
plt.bar(q,s,width=0.25)
plt.show()

# by default the ticks will appear 0.5 data points apart.
```

```python
q=range(4)
s=[23.5,25,26,28.5]
plt.bar(q,s,width=0.25)
plt.xticks([0,1,2,3])            # xticks(<sequence containing tick data points>,[<optional sequence containing tick labels])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
val=[[5.,25.,45.,20.],[4.,23.,49.,17.],[6.,22.,47.,19.]]
X=np.arange(4)
plt.bar(X+0.00,val[0],color='b',width=0.25,label='range1')
plt.bar(X+0.25,val[1],color='g',width=0.25,label='range2')
plt.bar(X+0.50,val[2],color='r',width=0.25,label='range3')
plt.legend(loc='upper left')                                     #  <matplotlib.pyplot>.legend(loc=<position number or string>)
plt.title("Multi-Range Bar Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
Data=[[5.,25.,45.,20.],[8.,13.,29.,27.],[9.,29.,27.,39.]]
X=np.arange(4)
plt.plot(X,Data[0],color='b',label='range1')
plt.plot(X,Data[1],color='g',label='range2')
plt.plot(X,Data[2],color='r',label='range3')
plt.legend(loc='lower right')                                     #  <matplotlib.pyplot>.legend(loc=<position number or string>)
plt.title("Multi-Range Line Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Saving a Figure

If you want to save a plot created using pyplot functions for later use or for keeping records, 
you can use savefig() to save the plot.

<matplotlib.pyplot>.savefig(<string with file name and path>)

```python
You can save the figures in popular formats like .pdf, .png, .eps and specify the format as file extensions

Also while specifying the path, use double slashes to supres special meaning of single slash character


plt.savefig("multibar.pdf")               # Save the plot in current directory

plt.savefig("C:\\data\\multibar.pdf")     # Save the plot at the given path

plt.savefig("C:\\data\\multibar.png")     # Save the plot at the given path in png format
```
