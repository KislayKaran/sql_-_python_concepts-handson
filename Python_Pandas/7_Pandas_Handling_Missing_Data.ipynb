# NA Values NULL or NaN or None

```python
import pandas as pd
import numpy as np

d1={'F_name':['Jivin',np.NaN,'Ruhi','Raunaq','Mike'],
    'L_name':['Arora',np.NaN,'Hyat','Singh','John'],
     'Age':[42,np.NaN,36,24,73],
     'Sex':['M',np.NaN,'F','M','F'],
     'Val_Before':[4,np.NaN,np.NaN,2,3],
     'Val_After':[25,np.NaN,np.NaN,62,70]}
df7=pd.DataFrame(d1)
df7
```

# (A) Detecting/Filtering Missing Data

isnull() --- to detect the missing values in a Pandas . It returns 'True' or 'False' for each values in the DF.
             will return True for the NaN values.
             
             df.isnull()

```python
df7.isnull()
```

notnull() --- if you want to filter data for series, which is not a missing value i.e. not-null data.

              series[series.notnull()]

# (B) Dropping Missing Values

You can either drop missing values or fill missing values.
To drop missing values you can use dropna() in fo;;owing 3 ways:
    
    (a)df.dropna()--- will drop all the rows that have NaN values
                      even row with only a Single NaN value.
        
    (b)df.dropna(how='all')--- will drop only those rows that have all NaN values.
    
    (c)df.dropna(axis=1,how='all')---will drop columns with all NaN values

```python
df7.dropna()
```

```python
df7.dropna(how='all')
```

# (C) Filling Missing Values

To avoid  losing the other not-NULL values in the same row data having null values in the same row data

having null values (to be removed). We need to fill the missing Data with some  appropriate value of our choice 

for this purpose we use fillna()

    df.fillna('value', inplace=True) 

the inplace argument will ensure that the filled missing values are reflected in the original df itself.

Using dictionary with fillna() to specify fill valuess for each column seperately.

```python
fillvalues={'F_name':'x',
    'L_name':'y',
     'Age':0,
     'Sex':'X',      
     'Val_Before':-999,
     'Val_After':-999}
pdf5=df7.fillna(fillvalues)
pdf5
```

```python
df7.fillna({'Age':0,'Sex':'x'})
```

# Filling Missing Values from Another DataFrame


Missing values of DataFrame1 (df1) will be filled from the corresponding cells of DataFrame2 (df2)

    df1.fillna(df2)

Condition here is that both the DFs should have similar structure, only their values can differ.
