# Combining DataFrames

# concat( ) Function

To combine DataFrmae along the rows or align the columns 

This method is useful if the two DataFrames having  similar structures.

pd.concat([df1,df2])                      DFs to be added are given as a sequence -------> List.

pd.concat([df1,df2],axis=1)               for column-concatation.

pd.concat([df1,df2],ignore_index= True)   To avoid repeatition index_values when concatenating row_index wise will Notice 
                                          new indexes are generated foer the resultant DataFrame from ) to n-1.

NOTE: 
      Column Concatenation shi=ould be avoided in case the Data structure (column_Labels) are excatly same.
    
      Instead we should go for row concatenation

```python
import pandas as pd
import numpy as np

d1={'Subject_id':[1,2,3,4,5],
    'F_name':['Jivin','Sarah','Ruhi','Raunaq','Mike'],
    'L_name':['Arora','Kapur','Hyat','Singh','John']}
df1=pd.DataFrame(d1)
df1     
```

```python
d2={'Subject_id':[6,7,8,9,10],
    'F_name':['Reet','Anil','Pramod','Jiya','Rudy'],
    'L_name':['Kaur','Sinha','Singh','Gopalan','Monga']}
df2=pd.DataFrame(d2)
df2 
```

```python
pd.concat([df1,df2])               # row-wise concatenation
```

```python
pd.concat([df1,df2],axis=1)     # column-wise concatenation
```

```python
pd.concat([df1,df2],ignore_index= True)
```

# join( ) Function

join() creats a new DFfrom two DFs by joining their 'Rows'


Types of Joins

(i) INNER Join : Take the rows with common indexes from both DataFrames and create new DataFrame.
    
    df1.join(df2, how='inner')
    
(ii) LEFT Join : Take all the rows from the Left DataFrame (first) and join with the only those rows
                 from the Second DataFrame that have common indexes to create a new DataFrame.
     
    df1.join(df2, how='left')    
    
    If skipped, the 'how' argument considers 'left' as the default value for the join.

(ii) RIGHT Join : Take all the rows from the Right DataFrame (second) and join with the only those rows
                  from the First DataFrame that have common indexes to create a new DataFrame.
     
    df1.join(df2, how='right')  

(iv) OUTER Join : Take all rows fro both DataFrames and create a new DataFrame . 
                  Showing missing values for non-matching indexes.
        
    df1.join(df2, how='outer')   

For non-matching indexes, missing values will be represented as NaN.

The join function expects the column names to be different in both the DataFrames
----> Common column names will lead to Value Error.

df1.join(df2,how='left') is same as df2.join(df1,how='right')

By default the 1st DF calling the join() is the left DF.
And the 2nd DF is the right DF.

```python
import pandas as pd
import numpy as np
```

```python
zone=['zoneA','zoneB','zoneC','zoneD']
ds1={'Target':[56000,70000,15000,60000],
     'Sales':[58000,68000,78000,61000]}
saledf1=pd.DataFrame(ds1,index=zone)
saledf1  
```

```python
zone=['zoneA','zoneB','zoneC','zoneD','zoneE','zoneF']
ds2={'Name':['Purv','Paschim','Kendriya','Dakshin','Uttar','Rural'],
    'Product':['Oven','AC','AC','Oven','TV','Tubewell']}
saledf2=pd.DataFrame(ds2,index=zone)
saledf2 
```

```python
newdf=saledf2.join(saledf1)         # If skipped, the 'how' argument considers 'left' as the default value for the join.
newdf
```

```python
newdf2=saledf2.join(saledf1,how='inner')    # Stores only the matching indexes' rows from both DFs. 
newdf2
```

```python
newdf3=saledf2.join(saledf1,how='right')    
newdf3
```

```python
newdf4=saledf2.join(saledf1,how='outer')    
newdf4
```

# Joining on a Column

By Default indexes of two DataFrames are matched but when we specify column of the first DataFrame with 'on' argument, 
then that column's values are matched with indexes of the second DataFrame.

    df1.join(df2,on ='column_name of df1')


```python
ds3={'Cust_id':[1,2,3,4,5,6],
    'Product':['Oven','AC','AC','Speaker','Tablet','Smartphoe']}
sales=pd.DataFrame(ds3)
sales 
```

```python
ds4={'P_id':[2,4,6],
    'State':['Delhi','Goa','Kerela']}
sales2=pd.DataFrame(ds4)
sales2 
```

```python
sales2.join(sales, on="P_id")       # Matching column values of P_id column in sales2 DF with indexes of the sales DF.
```

But the join() function can join only the left DataFrame's column values with the indexes of the right Dataframe.

It cannot join the values of columns  in both the DataFrames.

For that merge() function is used

when using the 'on' argument in join() provide 'lsuffix' and 'rsuffix' values, 
If the two DataFrames contain a column with same name.

    df1.join(df2, on= column_name, lsuffix='lsuffix_string', rsuffix='rsuffix_string')

    lsuffix for left Dataframe Column

    rsuffix for right Dataframe Column

So that it can create unique names for columns by adding suffixes, otherwise it will return an error.




```python
sales2.rename(columns={"P_id":"Cust_id"},inplace=True)    # Changing the P_id column from sales2 DF to Cust_id
```

```python
sales2
```

sales2.join(sales, on="Cust_id")    # this will give us error

```python
sales2.join(sales, on="Cust_id",lsuffix='_CDF',rsuffix='_SDF')
```

# merge( ) Function

To combine two DataFrames such that two rows with some common column value are merged together in the Final Result.

      pd.merge(df1,df2, on = 'field_name', how ='Left'/'Right'/'Inner'/'Outer')
      
on argument =  takes the column name found in both the dataFrames df1 and df2.
      
how argument =  specifies the type of merging which is actually a type of join.

NOTE:  If you skip the argument  on = 'field_name', then it will take any merge on common fields of two DFs..
       But you can explicitlyspecify the field on the basis of which you want to merge the two DataFrames.

```python
pd.merge(sales,sales2)
```

```python
pd.merge(sales,sales2,on='Cust_id')
```

```python
pd.merge(sales,sales2,on='Cust_id',how='left')
```

```python
dx={'Cust_id':[1,2,3,4,5],
    'F_name':['Jivin','Sarah','Ruhi','Raunaq','Mike'],
    'L_name':['Arora','Kapur','Hyat','Singh','John'],
    'email':['jvarora@kbc.com','shkapur@kkr.com','rihayat@srh.com','r_singh@csk.com','m_john@kix.com',]}
cust=pd.DataFrame(dx)
cust   
```

```python
dy={'Order_id':[1,2,3,4,5,6],
     'amount': [23234.56,62378.50,32124.00,8365.50,12325.50,12614.40],
     'Cust_id':[1,3,2,3,10,9]}
order=pd.DataFrame(dy)
order  
```

DFs merged on common field Cust_id and contains only the rows having matching cust_id values.

```python
res0=pd.merge(order,cust,on='Cust_id')
res0
```

DFs merged on common field Cust_id and contains all the rows from order DataFrame and from cust DataFrame,
only the rows having matching Cust_id.

```python
res1=pd.merge(order,cust,on='Cust_id',how='left')
res1
```

DFs merged on common field Cust_id and contains all the rows both from order and cust DataFrames.

```python
res2=pd.merge(order,cust,on='Cust_id',how='outer')
res2
```
