# Create a basic COUNTPLOT to get number of transaction for eachof the product line and change the figure size

```python
plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart)
```

```python
# Making it a Horizontal COUNTPLOT

sns.countplot(y='Product line',data=mart)
```

```python
# Adding hue to get the count on two categoriesi.e. Product Line and Gender

plt.figure(figsize=(15,5))

sns.countplot(x='Product line',hue='Gender',data=mart)
```

```python
# Using Different Color palette

plt.figure(figsize=(15,5))

sns.countplot(x='Product line',hue='Gender',data=mart,palette='plasma_r')
```

```python
# Changing Style using facecolor, linewidth and edge color

plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart,
             facecolor=(0,0,0,0),
             linewidth=5,
             edgecolor=sns.color_palette('dark',3))
```

# BARPLOT

In COUNTPLOTs we can only count the categorical values but in BARPLOTs we can get the Statistical Aggregation

like Mean, Median and others.

Error bars as well as columunars

A BarPlot represents an estimate of central tendency for a numeric variable with the height of each rectangle and 
provides some induction of the uncertainity around that estimate using error Bars.

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total',data=mart, order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],color='blue')
```

```python
# Changing the color using PALLETE Attribute

plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total',data=mart,hue='Gender', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=['Male','Female'],palette='cividis')
```

```python
# Using the saturation parameter

plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total',data=mart,hue='Gender', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=['Male','Female'],palette='winter', saturation=5)
```

```python
# Change DEFAULT AGGREGATION FUNCTION ( MEAN) method using ESTIMATOR PARAMETER

plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total',data=mart,hue='Gender', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=['Male','Female'],palette='cividis', estimator=sum)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total',data=mart,hue='Gender', order= ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=['Male','Female'],palette='Wistia', estimator= np.median)
```
