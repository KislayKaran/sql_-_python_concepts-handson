# VIOLIN PLOT

VIOLIN PLOT = BOXPLOT + KERNEL DENSITY

```python
seaborn.violinplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, bw='scott', 
                   cut=2, scale='area', scale_hue=True, gridsize=100, width=0.8, inner='box', split=False, 
                   dodge=True, orient=None, linewidth=None, color=None, palette=None, saturation=0.75, ax=None,
                   **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart.head()
```

```python
# Creating a VIOLINPLOT 

sns.violinplot(y='Total',data=mart)
```

```python
# Creating a VIOLINPLOT on two CATEGORICAL and one NUMERIC Variable and use split.
plt.figure(figsize=(10,5))
sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart)
```

```python
# Creating a VIOLINPLOT on two CATEGORICAL and one NUMERIC Variable and use split.

sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart, split=True)
```

```python
# Changing the box in the VIOLIN PLOT to horizontal lines

plt.figure(figsize=(15,5))

sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart, split=True, inner='quartile')

# Shows the 1st Quartile, 2nd Quartile, 3rd Quartile.
```

```python
# Draw line for each observation in a VIOLINPLOT

plt.figure(figsize=(15,5))

sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart, split=True, inner='stick', palette='afmhot')
```

```python
# Change the amount of smoothing using bw attribute

plt.figure(figsize=(15,5))

sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart, split=True, inner='quartile', palette='afmhot', bw=.6)
```

```python
# Cut out the Extreme Values

plt.figure(figsize=(15,5))

sns.violinplot(x='Payment',y='Total',hue='Gender', data=mart, split=True, 
               inner='quartile', palette='afmhot', bw=.6, cut=0)

# Removes the Max. and Min. extreme values.
```
