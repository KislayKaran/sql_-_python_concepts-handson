# STRIPPLOT

Jitter is the Feature in the STRIPPLOT using which the plotting of the repetitive values become Distinctive when plotted.

A Strip Plot is a graphical data analysis technique for summarizing a univariate data set

A strip plot can be drawn on its own, but it is also a good complement to a BOX or VIOLIN PLOT 
in cases where you want to show all observations  along with some representation of the underlying distribution.

```python
seaborn.stripplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, 
                  orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart.head()
```

```python
sns.stripplot(data=mart,y='Total')
```

```python
# Expand Markers in the STRIPPLOT

sns.stripplot(data=mart,y='Total', jitter=0.2)
```

```python
# Draw line around the points using LINEWIDTH

sns.stripplot(data=mart,y='Total', linewidth=0.5)

# around each of the point a line has ben drawn.
```

```python
# Including Third Categorical Value with HUE

plt.figure(figsize=(15,5))

sns.stripplot(data=mart,y='Total', x='Payment',hue='Gender', jitter=0.2, linewidth=0.5)
```

```python
# Seperate each Level of HUE using DODGE


sns.stripplot(data=mart,y='Total', x='Payment',hue='Gender', jitter=0.2, linewidth=0.5, dodge=True)
```

```python
# Draw the Strips on Top of VIOLIN PLOT
plt.figure(figsize=(10,5))

sns.violinplot(data=mart,y='Total', x='Payment', color='gray')
sns.stripplot(data=mart,y='Total', x='Payment', jitter=0.2)
```

```python
# Draw the Strips on Top of BOX PLOT
plt.figure(figsize=(10,5))

sns.boxplot(data=mart,y='Total', x='Payment', color='gray')
sns.stripplot(data=mart,y='Total', x='Payment', color='black')
```
