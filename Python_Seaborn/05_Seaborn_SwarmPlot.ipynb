# SWARMPLOT

To represent each of the Data Points on the Plot.

Same Data points overlap in Strip Plot,
In SwarmPlot all the Data Points are seperated and are not overlapping and clearly visible.

Mostly it is used for the small set of DataFrame.

```python
seaborn.swarmplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None,
                  color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart.head()
```

```python
sns.swarmplot(data=mart, y='Total')
```

```python
# Creating SWARMPLOT groupby Categories

plt.figure(figsize=(10,5))
sns.swarmplot(data=mart, y='Total',x='Payment')
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(data=mart, y='Total',x='Payment',hue='Gender')
```

```python
# Showing Hues Separately on Categorical axis

plt.figure(figsize=(10,5))
sns.swarmplot(data=mart, y='Total',x='Payment',hue='Gender', split=True)
```

```python
# Styling a SWARM : Changing the Marker, Size, Color, Edge Color

plt.figure(figsize=(15,5))
sns.swarmplot(data=mart, y='Total',x='Payment',marker='*', size=8 , color='green', edgecolor='black')
```

```python
# Overlay a SWARMPLOT on a BOXPLOT

plt.figure(figsize=(10,5))
sns.boxplot(data=mart, y='Total',x='Payment')
sns.swarmplot(data=mart, y='Total',x='Payment',color='black')
```

```python
# Overlay a SWARMPLOT on a VIOLINPLOT

plt.figure(figsize=(10,5))
sns.violinplot(data=mart, y='Total',x='Payment', inner=None, width=1)
sns.swarmplot(data=mart, y='Total',x='Payment',color='white')
```
