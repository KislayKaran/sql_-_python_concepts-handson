# HISTOGRAM PLOT

A Histogram is a classic Visualization tool that represents the distribution of one or more variables by counting
the number of Observations that fall within discrete bins.

This Function can normalize the static computed within each bin to estimate frequency, density or probability mass, 
and it can add a smooth curve obtained using a kernel density estimate, similar to kdeplot()

```python
seaborn.histplot(data=None, *, x=None, y=None, hue=None, weights=None, stat='count', bins='auto',
                 binwidth=None, binrange=None, discrete=None, cumulative=False, common_bins=True, 
                 common_norm=True, multiple='layer', element='bars', fill=True, shrink=1, kde=False, 
                 kde_kws=None, line_kws=None, thresh=0, pthresh=None, pmax=None, cbar=False, cbar_ax=None, 
                 cbar_kws=None, palette=None, hue_order=None, hue_norm=None, color=None,log_scale=None,
                 legend=True, ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart=mart[['Gender','Payment','Unit price', 'Quantity', 'Total', 'gross income']]
mart.head(10)
```

```python
# Create Histogram on X-axis or Y-axis

sns.histplot(data=mart, x='Total')
```

```python
sns.histplot(data=mart, y='Total')
```

```python
# Change the Bins Width using BinWidth argument
sns.histplot(data=mart, x='Total', binwidth=50)
```

```python
# Changing the no of Bins and Interval

plt.figure(figsize=(15,5))

sns.histplot(data=mart, x='Total', bins=np.arange(0,1100,50))

plt.xticks(np.arange(0,1100,50))
```

```python
# Combining with KDE

sns.histplot(data=mart, x='Total', kde=True)


# This Line shows the smooth density distribution
```

```python
# use a categorical variable in HUE and STACK it using MULTIPLE argument

plt.figure(figsize=(10,5))

sns.histplot(data=mart, x='Total', hue='Payment',multiple='stack')
```

```python
# Making it a Step/Poly using ELEMENT argument and Change the FILL

sns.histplot(data=mart, x='Total',element='step', fill=False)
```

```python
sns.histplot(data=mart, x='Total',element='poly')
```

```python
sns.histplot(data=mart, x='Total', hue='Payment',element='poly')
```

```python
# Use Categorical Variable and shrink it instead of Numerical Variable

sns.histplot(data=mart, x='Payment',shrink=0.7)
```

```python
sns.histplot(data=mart, x='Total', stat='probability')

# By Default stat is Count, ......else 'density', 'probability', 'frequency'
```

```python
# creating a BiVariate Histogram  ( for two Numeric variables)

sns.histplot(data=mart, x='Total', y='gross income')
```
