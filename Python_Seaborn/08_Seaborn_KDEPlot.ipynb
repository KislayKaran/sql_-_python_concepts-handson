```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart=mart[['Gender','Payment','Unit price', 'Quantity', 'Total', 'gross income']]
mart.head(10)
```

```python

sns.kdeplot(data=mart, x='Total')
```

```python
# Adjusting the Bandwidth using bw_adjust

sns.kdeplot(data=mart, x='Total', bw_adjust=0.2)
```

```python
# Create KDEPLOT for all numeric variables in the DataFrame


sns.kdeplot(data=mart)
```

```python
# Group the KDE on a Category Variable
sns.kdeplot(data=mart, x='Total', hue='Payment')
```

```python
# Stack KDE on a Category using MULTIPLE argument

sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

```python
# Using a Log Scaling to map the Variable in KDE

sns.kdeplot(data=mart, x='Total', log_scale=True)
```

```python
# Change Styling of Hued KDE using Linewidth, palette, alpha etc

sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', linewidth=0.5, palette='gnuplot', alpha=0.3)
```

```python
# Ceating a BiVariate KDE:
plt.figure(figsize=(10,5))

sns.kdeplot(data=mart, x='Unit price',y='gross income')
```

```python
# Group the BiVariate KDE on a Categorical Variable and show the Contours
plt.figure(figsize=(10,5))

sns.kdeplot(data=mart, x='Unit price',y='gross income', hue='Gender', fill=True, levels=5)
```

```python
plt.figure(figsize=(10,5))

sns.kdeplot(data=mart, x='Unit price',y='gross income', hue='Gender', fill=True, levels=5, thresh=0.2)
```
