# RUG PLOT

Plot Marginal distributions by drawing ticks along the x and y axes.

This Function is intended to complement other plots by showing the location of individual observations in 
an unobtrusive way.


```python
seaborn.rugplot(x=None, *, height=0.025, axis=None, ax=None, data=None,y=None, hue=None, palette=None, hue_order=None,
                hue_norm=None, expand_margins=True, legend=True, a=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart=mart[['Gender','Payment','Unit price', 'Quantity', 'Total', 'gross income']]
mart.head(10)
```

```python
plt.figure(figsize=(10,5))
sns.rugplot(data=mart, x='Unit price')
```

```python
# Create a RUGPLOT for two Variables

plt.figure(figsize=(10,5))
sns.rugplot(data=mart, x='Unit price', y='gross income')
```

```python
# Group it by a categorical Variable using HUE

plt.figure(figsize=(10,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue= 'Gender')
```

```python
# Change the Heights of the Rugs

plt.figure(figsize=(10,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue= 'Gender', height=0.15)
```

```python
# Combine with the KDE Plot
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue= 'Gender')
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue= 'Gender', fill=True)
```

```python
# Combine with the SCATTER Plot
plt.figure(figsize=(15,5))

sns.scatterplot(data=mart, x='Unit price', y='gross income', hue= 'Gender')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue= 'Gender')
```

```python
# Show RUGPLOT outside of the axis

plt.figure(figsize=(15,5))
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue= 'Gender', fill=True)
sns.scatterplot(data=mart, x='Unit price', y='gross income', hue= 'Gender')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue= 'Gender',height=-0.01, clip_on=False)
```
