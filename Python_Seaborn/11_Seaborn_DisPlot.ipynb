# DISPLOT

DISTRIBUTION PLOT: To create all the Distribution Category Plots (RUG, ECDF, KDEP, Histogram )
    It also allows to split the Plots based on the Categorical Value

```python
seaborn.displot(data=None, *, x=None, y=None, hue=None, row=None, col=None, weights=None, kind='hist', rug=False, 
                rug_kws=None, log_scale=None, legend=True, palette=None, hue_order=None, hue_norm=None, color=None,
                col_wrap=None, row_order=None, col_order=None, height=5, aspect=1, facet_kws=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart=pd.read_excel("C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/supermarket_sales.xlsx")
mart=mart[['Gender','Payment','Unit price', 'Quantity', 'Total', 'gross income', 'Branch']]
mart.head(10)
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True)
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height': 0.05}) 

#kde_kws={'bw_adjust':9}
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height': 0.05},hue='Payment')
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height': 0.05},hue='Payment', multiple='stack')
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height': 0.05},hue='Payment', multiple='stack', col='Gender')
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height': 0.05},hue='Payment', multiple='stack', col='Branch', row='Gender')
```

```python
sns.displot(data=mart, y='Total', kde=True, rug=True, rug_kws={'height': 0.05},hue='Payment', multiple='stack', col='Branch', row='Gender')
```

```python
sns.displot(data=mart, x='Total',kind='ecdf',hue='Payment', col='Branch', row='Gender')
```
