# JOINT PLOT

JOINT Plot draws a plot of Two variables with BiVariate and Univariate Graphs

Shows the marginal Distribution of X-axis (above) plots and the Y-Axis plots ( right)

1 Relational Plots + 2 Marginal Plots = JointPlot

```python
seaborn.jointplot(*, x=None, y=None, data=None, kind='scatter', color=None, height=6, ratio=5, space=0.2, 
                  dropna=False, xlim=None, ylim=None, marginal_ticks=False, joint_kws=None, marginal_kws=None,
                  hue=None, palette=None, hue_order=None, hue_norm=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

```python
# Change its kind to "Scatter| KDE| Hist|HEX|REG |RESID"

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='scatter')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='kde')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='hist')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='hex')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='reg')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='resid')
```

```python
# Grouping based on Categorical Value


sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```python

sns.jointplot(data=iris, x='sepal_length', y='petal_length', color= 'red')
```

```python
# Formating All and Individual Plots


sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='BuPu')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', joint_kws=dict(marker='+',color='green'),
                                                              marginal_kws=dict(color='blue', kde=True))
```

```python
# Adjust the height, ratio, space and show/hide the marginal_ticks

sns.jointplot(data=iris, x='sepal_length', y='petal_length', height=8)
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', height=8, ratio=4, space=1, marginal_ticks=True)
```

```python
# Plot KDE and RUG Plot on top of the JOINT Plot

l=sns.jointplot(data=iris, x='sepal_length', y='petal_length',joint_kws=dict(color='purple'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot, height=0.1, color='violet')
```
