# PAIR PLOT

Plots pairwise relationships in a dataset

Daigonal Section/Segment (Intersection of Same Variables): Zipper length of x-axis & Zipper Length of y-axis
                                                           Marginal Distribution is in the form of Histogram


Upper Triangle

Lower Triangle 


Different variables intersecting in the form of a relation ( Scatter Plot)
Plotting relation b/w a set of variables (numeric)

```python
seaborn.pairplot(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter',
                 diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, 
                 diag_kws=None, grid_kws=None, size=None)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
iris=sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
sns.pairplot(data=iris)
```

```python
# Changing the Diagonal Plot KIND to KDE, HIST (default) or NONE

sns.pairplot(data=iris, diag_kind='kde')
```

```python
# changing the non-diagonal plot KIND to Scatter, KDE HIST or Reg

sns.pairplot(data=iris, kind="reg")
```

```python
# Adding HUE to the PAIRPLOT

sns.pairplot(data=iris,hue='species')
```

```python
# Creating PAIR PLOT for specific list of variables(diag_kind=None)


sns.pairplot(data=iris, x_vars=['sepal_width','petal_width','petal_length'],y_vars=['sepal_length','sepal_width'], diag_kind=None)
```

```python
# Showing only the Lower Triangle using CORNER Argument

sns.pairplot(data=iris, corner=True)
```

```python
# Making Changes Specific to Diagonal and Non-Diagonal plots seperately

sns.pairplot(data=iris, diag_kws=dict(color='green', kde=True),
            plot_kws=dict(color='violet', marker=10, s=80))
```

```python
# Creating KDE plot on Top of the PAIR PLOT using map_lower or upper

l=sns.pairplot(data=iris, plot_kws=dict(color='red'))
l.map_upper(sns.kdeplot)
l.map_lower(sns.kdeplot)
```
