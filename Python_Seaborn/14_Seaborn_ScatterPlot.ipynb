# SCATTER PLOT

A Diagram which shows the relationship b/w two variables by plotting the points/dots.

```python
seaborn.scatterplot(*, x=None, y=None, hue=None, style=None, size=None, data=None, palette=None, hue_order=None, 
                    hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, 
                    x_bins=None, y_bins=None, units=None, estimator=None, ci=95, n_boot=1000, alpha=None,
                    x_jitter=None, y_jitter=None, legend='auto', ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv('C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/mart_train_sample_.csv')
```

```python
mart.head()
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales')
```

```python
# Grouping basis on a categorical variable using HUE

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', hue='Outlet_Size')
```

```python
# Grouping basis on a categorical variable using STYLE and Styling Markers as well

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size')
```

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size',
                                                    markers={"High":"^","Small":"v", "Medium":"o"})
```

```python
# Grouping on basis of a categorical variable using HUE & Style both together
plt.figure(figsize=(10,5))

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size', hue='Outlet_Size',
                                                    markers={"High":"^","Small":"v", "Medium":"o"})
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size', hue='Outlet_Location_Type',
                                                    markers={"High":"^","Small":"v", "Medium":"o"})

# avoid using two different variables in style and hue.....
# instead use the same
```

```python
# Grouping on basis of a Numerical variable using HUE and using PALETTE

plt.figure(figsize=(10,5))

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', hue='Outlet_Size',size='Outlet_Size',sizes=(20,200),
                            legend='full', palette="Purples_r",markers={"High":"^","Small":"v", "Medium":"o"})
          
```

```python
# Changing the Marker Size with 's' argument

plt.figure(figsize=(10,5))

sns.scatterplot(data=mart, x='Item_MRP', y='Sales', s=400, color='red', edgecolor='black')
```
