# LINE PLOT

Line Plot shows the relation b/w the two variables.

Showcases how the value of variable changes over the time.

Majorly used in time serie analysis

```python
seaborn.lineplot(*, x=None, y=None, hue=None, size=None, style=None,data=None, palette=None, hue_order=None,
                 hue_norm=None, sizes=None, size_order=None, size_norm=None, dashes=True, markers=None,
                 style_order=None, units=None, estimator='mean', ci=95, n_boot=1000, seed=None, sort=True,
                 err_style='band', err_kws=None, legend='auto', ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv('C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/mart_train_sample_.csv')
mart.head()
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales')

# The shaded region shows the Bootstraped Confidence Interval , whilr the Line represents the Mean of the Sales
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None)

# without Confidence Interval
```

```python
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator=sum)
```

```python
# with estimator=None, we can see each of the Data Ponts by the year

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator=None)
```

```python
# Testing with Different Confidence Intervals with different Number of Bootstraping

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci='sd')

# Standard Deviation sd

# By Default the CI is 95 % 
```

```python
# By Default no of BootStrap is 1000

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', n_boot=20)
```

```python
# Groupig usinng HUE and using different Palette
plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', palette='rocket_r')
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', palette=['red','green','blue'])
```

```python
#  rouping using Style, use different Marker and  ashe

plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', style='Outlet_Size', markers=True)
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, hue='Outlet_Size', style='Outlet_Size',
             markers=True, dashes=False)
```

```python
# grouping using Size

plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, size='Outlet_Size',sizes=(.5,5) )
```

```python
# using Different Semantic Styling Parameters on same or Different Variables

plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None,hue='Outlet_Size', style='Outlet_Size', markers=True)
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None,hue='Outlet_Size', style='Tier', markers=True)
```
