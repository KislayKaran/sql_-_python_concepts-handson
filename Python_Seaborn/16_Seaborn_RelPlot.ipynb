# REL PLOT

Relational Plot : It gives the option to switch b/w Line Plot and Scatter Plot
   
   col_wrap  ( is an important Attribute)

```python
seaborn.relplot(*, x=None, y=None, hue=None, size=None, style=None,data=None, row=None, col=None, col_wrap=None,
                row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None, 
                size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='auto',
                kind='scatter', height=5, aspect=1, facet_kws=None, units=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv('C:/Users/Kislay Karan/Downloads/New folder/New folder/Python_Seaborn/mart_train_sample_.csv')
mart.head()
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100)

# by default it is Scatter Plot
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', kind='line', ci=None)
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100, hue='Outlet_Size', col='Tier')
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100, hue='Outlet_Size', col='Item_Type')
```

```python
# To solve the above problem we can specify the no of columns you want in each row we can use col_wrap

sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100, hue='Outlet_Size', col='Item_Type', col_wrap=3)
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100,
            hue='Outlet_Size',col='Outlet_Size', row='Tier')
```

```python
sns.relplot(data=mart, x='Outlet_Year', y='Sales', s=100,
            hue='Outlet_Size',palette=['green','red','blue'])
```
