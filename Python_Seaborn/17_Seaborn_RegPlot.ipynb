# REG PLOT

Regression Plot  : It Plots Data and a linear regression model fit.

 Finds the best fit Linear Regression Line which helps to predict the amount of change in Y on one unit change in X.

From the Scatter Plot....

It just adds a best fit regression line which helps in predicting the value change on Y-axis which occurs 
with each unit change in X-axis.

And the shaded region shows the Confidence Interval Region which is built on the basis of Bootstraping. 

```python
seaborn.regplot(*, x=None, y=None, data=None, x_estimator=None, x_bins=None,x_ci='ci', scatter=True, 
                fit_reg=True, ci=95, n_boot=1000, units=None, seed=None, order=1, logistic=False, 
                lowess=False, robust=False, logx=False, x_partial=None, y_partial=None, truncate=True, 
                dropna=True, x_jitter=None, y_jitter=None, label=None, color=None, marker='o', scatter_kws=None, 
                line_kws=None, ax=None)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# Change the Styling of RegPlot, only Scatter, Show/hide the ci, change the ci value,change n_boot

sns.regplot(data=tips, x='total_bill', y='tip', marker='+',
            line_kws=dict(color='blue', linestyle='--'), scatter_kws=dict(s=100, color='green', alpha=0.5))
```

```python
sns.regplot(data=tips, x='total_bill', y='tip',ci=None, n_boot=500,
            line_kws=dict(color='blue', linestyle='--'), scatter_kws=dict(s=100, color='green', alpha=0.5))
```

```python
# Statistical model in Seaborn

sns.regplot(data=tips, x='total_bill', y='tip', order=1)

# Binary dependent variables to make the Line Flexible to make it more linear (Straight)
```
