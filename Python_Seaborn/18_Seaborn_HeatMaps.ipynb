# HEATMAPs

A Heatmap is a Graphical representation of data where values are depicted by color.

```python
Before creating the HeatMaps its important that the Data must be in the forn PIVOT or Correlation
```

```python
seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g',
                annot_kws=None, linewidths-0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False,
                xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel(r'C:\Users\Kislay Karan\Downloads\New folder\New folder\Python_Seaborn\mart_linePlot_.xlsx')
```

```python
mart.columns=mart.columns.str.lower()
mart.head(10)
```

```python
mart.shape
```

```python
martPiv=mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
martPiv
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))

sns.heatmap(martPiv, annot=True, fmt='.0f', annot_kws=dict(size=15, weight='bold'), linewidths=0.5, linecolor='black') 
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))

sns.heatmap(martPiv, annot=True, fmt='.0f', annot_kws=dict(size=15, weight='bold'),
            linewidths=0.5, linecolor='black', cmap='OrRd_r')
```

```python
# Data with Correlation, Value ranges b/w -1 to +1.

mart.corr()
```

```python
sns.heatmap(mart.corr())
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(mart.corr(), vmin=-1, vmax=1, center=0)
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(mart.corr(), vmin=-1, vmax=1, center=0,cmap='OrRd_r', annot= True, fmt='.1f',
                                                    annot_kws=dict(size=15, weight='bold'),linewidths=0.5, linecolor='black')
           
```
