# CLUSTER MAP

Used in clustering and Cluster analysis

It is used to plot a matrix dataset as hiearchially clustered Heatmap.

Cluster Map groups those columns that have very identical values.

```python
seaborn.clustermap(data, *, pivot_kws=None, method='average', metric='euclidean', z_score=None, standard_scale=None,
                   figsize=(10, 10), cbar_kws=None, row_cluster=True, col_cluster=True, row_linkage=None,
                   col_linkage=None, row_colors=None, col_colors=None, mask=None, dendrogram_ratio=0.2, 
                   colors_ratio=0.03, cbar_pos=(0.02, 0.8, 0.05, 0.18), tree_kws=None, **kwargs)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
```

```python
mart=pd.read_excel(r'C:\Users\Kislay Karan\Downloads\New folder\New folder\Python_Seaborn\mart_linePlot_.xlsx')
mart.columns=mart.columns.str.lower()
mart.head(10)
```

```python
martPiv=mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
martPiv
```

```python
sns.clustermap(martPiv)
```

```python
# Specfying the Clusters

sns.clustermap(martPiv, col_cluster=False)

# No Column Cluster
```

```python
# Only Column cluster

sns.clustermap(martPiv, row_cluster=False)
```

```python
# Features used in HeatMap

sns.clustermap(martPiv, row_cluster=False, annot=True, fmt='.0f')
```

```python
sns.clustermap(martPiv, row_cluster=False, annot=True, fmt='.0f', z_score=1)
```

```python
sns.clustermap(martPiv, row_cluster=False, annot=True, fmt='.0f', standard_scale=1, linewidth=.5)
```
