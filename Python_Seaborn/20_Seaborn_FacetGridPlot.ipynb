```python
seaborn.FacetGrid(self, data, *, row=None, col=None, hue=None, col_wrap=None, sharex=True, sharey=True, 
                  height=3, aspect=1, palette=None, row_order=None, col_order=None, hue_order=None, 
                  hue_kws=None, dropna=False, legend_out=True, despine=True, margin_titles=False, xlim=None,
                  ylim=None, subplot_kws=None, gridspec_kws=None, size=None)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_excel(r'C:\Users\Kislay Karan\Downloads\New folder\New folder\Python_Seaborn\mart_linePlot_.xlsx')
mart.columns=mart.columns.str.lower()
mart.head(10)
```

```python
mart.outlet_size.unique()
```

```python
mart.outlet_location_type.unique()
```

```python
l=sns.FacetGrid(mart,col='outlet_size')
l.map_dataframe(sns.histplot, x='sales')
```

```python
l=sns.FacetGrid(mart,col='outlet_size')
l.map_dataframe(sns.kdeplot,x='sales')
```

```python
l=sns.FacetGrid(mart,col='outlet_size')
l.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

```python
l=sns.FacetGrid(mart,col='outlet_size')
l.map_dataframe(sns.scatterplot, x='outlet_year', y='sales', hue='outlet_location_type',
               marker='+', alpha=0.5, color='red', s=100)
```

```python
plt.figure(figsize=(15,5))
l=sns.FacetGrid(mart,col='outlet_size', row='outlet_location_type', sharey=False,ylim=(0,1000))
l.map_dataframe(sns.histplot, x='sales')


l.set_axis_labels('sales','count')
l.set_titles(col_template='{col_name} Size',row_template='{row_name} Type')
```
